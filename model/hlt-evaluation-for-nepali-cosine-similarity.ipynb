```python
import torch
from torch.nn.functional import cosine_similarity

from sentence_transformers import SentenceTransformer, models
```

```python
"""
model_base
model_finetuned_QQP 
model_finetuned_snli
model_finetuned_snli_nepali
"""
```

```python
word_embedding_model = models.Transformer('google/muril-base-cased')
pooling_model = models.Pooling(word_embedding_model.get_word_embedding_dimension())
model_base = SentenceTransformer(modules=[word_embedding_model, pooling_model])
```

```python
word_embedding_model = models.Transformer('Yunika/muril-base-sentence-transformer')
pooling_model = models.Pooling(word_embedding_model.get_word_embedding_dimension())
model_finetuned_QQP = SentenceTransformer("Yunika/muril-base-sentence-transformer", modules=[word_embedding_model, pooling_model])
```

```python
word_embedding_model = models.Transformer('sbastola/muril-base-cased-sentence-transformer-snli')
pooling_model = models.Pooling(word_embedding_model.get_word_embedding_dimension())
model_finetuned_snli = SentenceTransformer("sbastola/muril-base-cased-sentence-transformer-snli", modules=[word_embedding_model, pooling_model])
```

```python
word_embedding_model = models.Transformer('sbastola/muril-base-cased-sentence-transformer-snli-nepali')
pooling_model = models.Pooling(word_embedding_model.get_word_embedding_dimension())
model_finetuned_snli_nepali = SentenceTransformer("sbastola/muril-base-cased-sentence-transformer-snli-nepali", modules=[word_embedding_model, pooling_model])
```

```python
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/sentence-compression-nepali/train.csv")
```

```python
df.head()
```

```python
df.columns = [["id", "first", "second", "a", "b", "c", "d"]]
```

```python
df = df[["first", "second"]]
```

```python
df.head()
```

```python
dataset = df.values.tolist()
```

```python
text1, text2 = [], []
for x, y in dataset[:1000]:
    text1.append(x)
    text2.append(y)
```

```python
embedding_1 = model_finetuned_QQP.encode(text1)
embedding_2 = model_finetuned_QQP.encode(text2)
embedding_1 = torch.tensor(embedding_1)
embedding_2 = torch.tensor(embedding_2)
sim_qqp = cosine_similarity(embedding_1, embedding_2)
```

```python
embedding_1 = model_finetuned_snli_nepali.encode(text1)
embedding_2 = model_finetuned_snli_nepali.encode(text2)
embedding_1 = torch.tensor(embedding_1)
embedding_2 = torch.tensor(embedding_2)
sim_nepali = cosine_similarity(embedding_1, embedding_2)
```

```python
average_sim_base = sim_qqp.mean()
average_sim_nepali = sim_nepali.mean()
```
